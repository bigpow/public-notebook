```python
from __future__ import print_function
import requests


def word_counters(text):
    """Count words and returns a dict."""
    counts = {}
    for word in text:
        word = word.replace(",", "").replace(".", "")
        if word not in counts:
            counts[word] = 1
        else:
            counts[word] = counts[word] + 1
    return counts

def print_sort_dict(data_dict, sort='keys', reverse=False):
    """Print a dict sorted by key or value."""
    if sort == "keys":
        # sorted by keys
        for key, val in sorted(data_dict.iteritems(), reverse=reverse):
            print("{:20}{:>5}".format(key, val))
    elif sort == "values":
        # sorted by values
        sort_key = lambda (k,v): (v,k)
        for key, val in sorted(data_dict.iteritems(), key=sort_key, reverse=reverse):
            print("{:20}{:>5}".format(key, val))
```
