```python
from __future__ import print_function
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("c:\\projects\\notebooks\\chickenducknutrients.csv")
    
```

```python
df
```

```python
df['Difference'] = df['Duck']/df['Chicken']
df['Difference'] = [ round(elem, 2) for elem in df['Difference'] ]
```

```python
df
```

```python
fig = plt.figure(dpi=100, figsize=(10, 5))
ax = fig.add_subplot(1,1,1)
ax.barh(df['Nutrient'], df['Difference'], label="Difference", align='center', alpha=0.5)
ax.set_xticklabels([ str(elem)+'X' for elem in df['Difference'] ], rotation=45)
fig.legend()
fig.tight_layout()
```
